```python
# Extracting data and creating a DataFrame
def concat_lc_objs(data_for_df,some_lcs):
    not_found = []
    for d in some_lcs:
        lc = d['candidates']  # This gets the nested dictionary
        if len(lc) > 0:
            lc_df = pd.DataFrame(lc, columns=['jd','magpsf','sigmapsf','magzpsci', 'fid','isdiffpos'])
            lc_df = lc_df[lc_df['magpsf'].notna()]
            lc_df.loc[:,'objId'] = d['objectId']
            data_for_df.append(lc_df)
        else:
            not_found.append(d['objectId'])
    return not_found
```

```python
tns_sn['ZTFID'].shape
```

```python
#get rid of nonvalid IDs
tns_sn_clean = tns_sn[~pd.isnull(tns_sn["ZTFID"])]
tns_sn_clean.shape
```

```python
#ensure the IDs have the proper format (start with ztf)
tns_sn_clean = tns_sn_clean[tns_sn_clean["ZTFID"].str.startswith("ZTF")]
tns_sn_clean.shape
```

```python
#ok let's do the whole query
id_list = tns_sn_clean['ZTFID'].to_list()
id_list
```

```python
print(len(id_list))
n_batches = len(id_list)/50
n_batches = int(n_batches)
batch_size = 50
```

```python
def make_batches(id_list):
    batches = []    
    c=0
    for b in range(n_batches-1):
        li = b+c
        hi = b+c+batch_size+1
        batch = id_list[li:hi]
        batches.append(batch)
        c+=batch_size
    return batches
    
batches = make_batches(id_list)
print(batches)
print(len(batches))
```

```python
data_for_lc_df = []
not_found = []

for batch in batches:
    batch_lcs = L.lightcurves(batch)
    not_found.append(concat_lc_objs(data_for_lc_df,batch_lcs))

print(not_found)
print(data_for_lc_df)
```
